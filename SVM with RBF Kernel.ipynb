```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import svm
from sklearn.datasets import make_circles

# 生成非线性可分的数据
X, y = make_circles(n_samples=100, factor=0.3, noise=0.1)

# 使用RBF核进行SVM分类
clf = svm.SVC(kernel='rbf', C=1.0, gamma='auto')
clf.fit(X, y)

# 创建一个网格来绘制决策边界
xx, yy = np.meshgrid(np.linspace(-1.5, 1.5, 500), np.linspace(-1.5, 1.5, 500))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# 绘制数据点和决策边界
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap=plt.cm.PuBu)
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='darkred')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k')
plt.title('Non-linear SVM with RBF kernel')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()

# 3D绘图
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 扩展数据为3D
r = np.exp(-(X ** 2).sum(1))
ax.scatter(X[:, 0], X[:, 1], r, c=y, cmap=plt.cm.Paired, edgecolors='k')

# 绘制3D决策边界
zz = np.exp(-(np.c_[xx.ravel(), yy.ravel()] ** 2).sum(1))
zz = zz.reshape(xx.shape)
ax.plot_surface(xx, yy, zz, rstride=1, cstride=1, color='grey', alpha=0.3)

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('RBF Kernel Feature')
plt.title('3D plot of Non-linear SVM with RBF kernel')
plt.show()
```
